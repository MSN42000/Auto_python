### Exercice 5 : Écriture dans un fichier CSV
Écrivez un programme Python qui demande à l'utilisateur de saisir le nom, l'âge et le pays d'origine de trois personnes, puis écrit ces informations dans un fichier CSV nommé "personnes.csv".


```python
import csv

personnes = []
for i in range(3):
    print(f"\nPersonne {i+1}:")
    nom = input("Nom: ")
    age = input("Age: ")
    pays = input("Pays d'origine: ")
    print([nom, age, pays])
    personnes.append([nom, age, pays])

with open("../Data/personnes.csv", "w", newline='') as fichier:
    writer = csv.writer(fichier)
    writer.writerow(["Nom", "Age", "Pays"])
    writer.writerows(personnes)
```

### Exercice 7 : Recherche dans un fichier
Écrivez un programme Python qui demande à l'utilisateur de saisir un mot-clé, puis recherche ce mot-clé dans un fichier texte nommé "texte.txt". Affichez toutes les lignes contenant le mot-clé ainsi que le nombre total d'occurrences.

```python
mot_cle = input("Mot-clé: ")

with open("../Data/texte.txt", "r") as fichier:
    lignes = fichier.readlines()
    lignes_trouvees = []
    compteur = 0
    for ligne in lignes:
        if ligne.find(mot_cle) != -1:
            lignes_trouvees.append(ligne.strip())
            compteur += 1

for ligne in lignes_trouvees:
    print(ligne)

print(f"\nNombre total d'occurrences du mot-clé \"{mot_cle}\" est : {compteur}")
```

### Exercice 9 : Conversion de données CSV
Écrivez un programme Python qui lit les données d'un fichier CSV nommé "donnees.csv", effectue des calculs sur ces données (par exemple, calcul de la moyenne, du total, etc.), puis écrit les résultats dans un nouveau fichier CSV nommé "resultats.csv".

```python
import csv

# Ouvrir les fichiers d'entrée et de sortie
with open("../Data/donnees.csv", "r") as fichier_entree, open("../Data/resultats.csv", "w", newline="") as fichier_sortie:
    lecteur_csv = csv.reader(fichier_entree)
    ecrivain_csv = csv.writer(fichier_sortie)
    
    # Lire l'en-tête et ignorer
    next(lecteur_csv)
    
    # Lire les données du fichier
    donnees = [ligne for ligne in lecteur_csv]
    
    # Calculer la somme de la deuxième colonne (index 1)
    somme = sum(float(ligne[1]) for ligne in donnees if ligne[1].replace('.', '', 1).isdigit())
    
    # Calculer la moyenne de la troisième colonne (index 2)
    moyenne = sum(float(ligne[2]) for ligne in donnees if ligne[2].replace('.', '', 1).isdigit()) / len(donnees) if len(donnees) > 0 else 0
    
    # Écrire les résultats dans le fichier de sortie
    ecrivain_csv.writerow(["Somme", "Moyenne"])
    ecrivain_csv.writerow([somme, moyenne])
    
    print("Les résultats ont été enregistrés dans 'resultats.csv'.")
```
